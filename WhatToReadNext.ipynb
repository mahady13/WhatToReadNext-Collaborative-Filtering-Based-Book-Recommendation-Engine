```python
import numpy as np
import pandas as pd
import scipy
import sklearn.neighbors as neighbors
```

```python
books=pd.read_csv("Books.csv")
ratings=pd.read_csv("Ratings.csv")
users=pd.read_csv("Users.csv")
```

```python
books.head()
```

```python
ratings.head()
```

```python
books_with_ratings=books.merge(ratings,on='ISBN')
```

```python
books_with_ratings.head()
```

```python
x=books_with_ratings.groupby('User-ID').count()['Book-Rating']>200
special_users=x[x].index
special_users
```

```python
rating_by_best_users=books_with_ratings[books_with_ratings['User-ID'].isin(special_users)]
```

```python
rating_by_best_users.head()
```

```python
y=rating_by_best_users.groupby('Book-Title').count()['Book-Rating']>=50
popular_books=y[y].index
```

```python
popular_books
```

```python
best_rated_by_best_users=rating_by_best_users[rating_by_best_users['Book-Title'].isin(popular_books)]
```

```python
best_rated_by_best_users.head()
```

```python
best_rated_by_best_users.shape
```

```python
pt=best_rated_by_best_users.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
pt.fillna(0,inplace=True)
```

```python
index=np.where(pt.index=="Animal Farm")[0][0]
index
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(pt)
similarity
```

```python
sorted(list(enumerate(similarity[254])),key=lambda x:x[1],reverse=True)[1:6]
```

```python
def recommend_books(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")['Book-Author'].values))
        data.append(item)

    return data
```

```python
recommend_books('Animal Farm')
```
